```python
import pandas as pd
import numpy as np
vdjdb = pd.read_csv('vdjdb_TRA_pairs.csv')
McPAS = pd.read_csv('McPAS_alpha_pairs.csv')
IEDB = pd.read_csv('IEDB_alpha_pairs_f.csv')
TBAdb = pd.read_csv('TRAdb_alpha_pairs_f.csv')
```

```python
vdjdb['Score'].value_counts()
vdjdb_2_3 = vdjdb.loc[vdjdb["Score"] >= 2 ]
vdjdb_2_3 = vdjdb_2_3.drop(columns=['Score'])
vdjdb_2_3 = vdjdb_2_3.drop_duplicates()

IEDB = IEDB.rename(columns={"Chain 1 - CDR3 Curated": "CDR3", "Epitope - Name": "Epitope"})
IEDB = IEDB.drop_duplicates()

TBAdb = TBAdb.rename(columns={"CDR3.alpha.aa": "CDR3", "Antigen.sequence": "Epitope"})
TBAdb = TBAdb.drop(columns=['Grade'])
TBAdb = TBAdb.drop_duplicates()

McPAS = McPAS.drop(columns=['Species', 'MHC'])
McPAS = McPAS.rename(columns={"CDR3.alpha.aa": "CDR3", "Epitope.peptide": "Epitope"})
McPAS = McPAS.drop_duplicates()

cols = list(vdjdb_2_3.columns)
cols
TBAdb = TBAdb[cols]

alpha_pairs = pd.concat([vdjdb_2_3, McPAS, IEDB, TBAdb])
alpha_pairs = alpha_pairs.drop_duplicates()
alpha_pairs[['Epitope','Extra']] = alpha_pairs['Epitope'].str.split('+',expand=True)
alpha_pairs[['Epitope','Extra1']] = alpha_pairs['Epitope'].str.split(' ',expand=True)
alpha_pairs = alpha_pairs[~alpha_pairs['Epitope'].str.contains('\d+')]
alpha_pairs = alpha_pairs[alpha_pairs['Epitope'].str[0].str.isupper()]
alpha_pairs = alpha_pairs.drop(columns=['Extra'])
alpha_pairs = alpha_pairs.drop(columns=['Extra1'])
alpha_pairs.to_csv(r'E:\Microsoft VS Code\alpha_pairs_dataset.csv',  index=False, header=True)

vdjdb_2_3['TCRepitope'] = vdjdb_2_3['CDR3']+' - '+ vdjdb_2_3['Epitope']
McPAS['TCRepitope'] = McPAS['CDR3']+' - '+ McPAS['Epitope']
IEDB['TCRepitope'] = IEDB['CDR3']+' - '+ IEDB['Epitope']
TBAdb['TCRepitope'] = TBAdb['CDR3']+' - '+ TBAdb['Epitope']

vdjdb_list = vdjdb_2_3["TCRepitope"].values.tolist()
McPAS_list = McPAS["TCRepitope"].values.tolist()
IEDB_list = IEDB["TCRepitope"].values.tolist()
TBAdb_list = TBAdb["TCRepitope"].values.tolist()

pip install venn
from venn import venn
datas = {
    "Vdjdb": set(vdjdb_list),
    "McPAS": set(McPAS_list),
    "IEDB": set(IEDB_list),
    "TBAdb": set(TBAdb_list),
}
plt.figure(figsize=(64, 8))
venn(datas)
plt.savefig('myplot.png')
```
